```python
# This notebook calculates and plots different revolution times and acceleration differences for a person of 2 meters
# with regards to artificial gravitation.
import math
def revTime(radii, baseAcc=9.81):
    time = math.sqrt(4*math.pi**2*radii/baseAcc)
    return time

# Simplified a = ((4*pi^2*(r-2))/(revTime(r))^2) 
def accAtRadiiMinusTwo(radii, baseAcc = 9.81):
    acc = baseAcc - 2*baseAcc/radii
    return acc

# Quotient of previous function and baseAcc
def accRadiiMinusTwoDivBaseAcc(accMinusTwo, baseAcc=9.81):
    return accMinusTwo/baseAcc
```

```python
import numpy as np
# Create an array of 141 numbers to represent different radiis of the spacecraft
radiis = np.linspace(10, 250, 241, dtype=int)
accAtRadii = []
for rad in radiis: 
    accAtRadii.append(accRadiiMinusTwoDivBaseAcc(accAtRadiiMinusTwo(rad)))
accAtRadii = np.array(accAtRadii)
```

```python
import matplotlib.pyplot as plt

fig = plt.figure()
fig.patch.set_facecolor("xkcd:white")
plt.plot(radiis, accAtRadii)

plt.xlabel("Radie (m)")
plt.ylabel("AccAtRadiiMinusTwo/AccAttRadii")
plt.grid()
```

```python
fig.savefig("artificialGravity.png", dpi=400, bbox_inches="tight")
```

```python
revTimeAtRadii = []
for rad in radiis:
    revTimeAtRadii.append(revTime(rad))
revTimeAtRadii = np.array(revTimeAtRadii)
```

```python
fig = plt.figure()
fig.patch.set_facecolor("xkcd:white")
plt.plot(radiis, revTimeAtRadii)

plt.xlabel("Radie (m)")
plt.ylabel("Omloppstid (s)")
plt.grid()
```
